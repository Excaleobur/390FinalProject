```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#importing data
# jump = pd.read_csv('C:/Users/Cassie G/Documents/390FinalProject-1/allJump.csv')
# walk = pd.read_csv('C:/Users/Cassie G/Documents/390FinalProject-1/allWalk.csv')
# both = pd.read_csv('C:/Users/Cassie G/Documents/390FinalProject-1/datasetOverall.csv')

jump = pd.read_csv('allJump.csv')
walk = pd.read_csv('allWalk.csv')
both = pd.read_csv('datasetOverall.csv')

#renaming the columns in the dataset so it is easier to call - column names (time, x,y,z,abs) are the same for all datasets
jump = jump.rename(columns={"Time (s)" : "time", "Linear Acceleration x (m/s^2)" : "x", 
                     "Linear Acceleration y (m/s^2)" : "y", "Linear Acceleration z (m/s^2)" : "z", 
                     "Absolute acceleration (m/s^2)" : "abs"})

walk = walk.rename(columns={"Time (s)" : "time", "Linear Acceleration x (m/s^2)" : "x", 
                     "Linear Acceleration y (m/s^2)" : "y", "Linear Acceleration z (m/s^2)" : "z", 
                     "Absolute acceleration (m/s^2)" : "abs"})

both = both.rename(columns={"Time (s)" : "time", "Linear Acceleration x (m/s^2)" : "x", 
                     "Linear Acceleration y (m/s^2)" : "y", "Linear Acceleration z (m/s^2)" : "z", 
                     "Absolute acceleration (m/s^2)" : "abs"})
```

```python
#plotting x and z variables within each dataset to compare
jumpXZ = plt.scatter(jump.loc[:,["x"]], jump.loc[:,["z"]])

plt.title("Jumping Data")
plt.xlabel("Linear Acceleration x")
plt.ylabel("Linear Acceleration z")

plt.show()
```

```python
walkXZ = plt.scatter(walk.loc[:,["x"]], walk.loc[:,["z"]])

plt.title("Walking Data")
plt.xlabel("Linear Acceleration x")
plt.ylabel("Linear Acceleration z")

plt.show()
```

```python
bothXZ = plt.scatter(both.loc[:,["x"]], both.loc[:,["z"]])

plt.title("Both Data")
plt.xlabel("Linear Acceleration x")
plt.ylabel("Linear Acceleration z")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


jump = pd.read_csv('allJump.csv')
walk = pd.read_csv('allWalk.csv')
both = pd.read_csv('datasetOverall.csv')

#renaming the columns in the dataset so it is easier to call - column names (time, x,y,z,abs) are the same for all datasets
jump = jump.rename(columns={"Time (s)" : "time", "Linear Acceleration x (m/s^2)" : "x", 
                     "Linear Acceleration y (m/s^2)" : "y", "Linear Acceleration z (m/s^2)" : "z", 
                     "Absolute acceleration (m/s^2)" : "abs"})

walk = walk.rename(columns={"Time (s)" : "time", "Linear Acceleration x (m/s^2)" : "x", 
                     "Linear Acceleration y (m/s^2)" : "y", "Linear Acceleration z (m/s^2)" : "z", 
                     "Absolute acceleration (m/s^2)" : "abs"})

both = both.rename(columns={"Time (s)" : "time", "Linear Acceleration x (m/s^2)" : "x", 
                     "Linear Acceleration y (m/s^2)" : "y", "Linear Acceleration z (m/s^2)" : "z", 
                     "Absolute acceleration (m/s^2)" : "abs"})


#This code creates a figure with 3 subplots arranged vertically, and creates bar graphs for the "x", "y", and "z" columns of the jump2 dataframe.

fig, ax = plt.subplots(3, 1, figsize=(10, 8))
ax[0].bar(jump['time'], jump['x'], width=0.5)
ax[0].set_title('Jump - X vs Time')
ax[0].set_xlabel('time')
ax[0].set_ylabel('x')

ax[1].bar(jump['time'], jump['y'], width=0.5)
ax[1].set_title('Jump - Y vs Time')
ax[1].set_xlabel('time')
ax[1].set_ylabel('y')

ax[2].bar(jump['time'], jump['z'], width=0.5)
ax[2].set_title('Jump - Z vs Time')
ax[2].set_xlabel('time')
ax[2].set_ylabel('z')


plt.tight_layout()
plt.show()

# create bar graphs for x, y, and z columns for walk data
# fig, ax = plt.subplots(3, 1, figsize=(10, 8))
# ax[0].bar(walk['time'], walk['x'], width=0.5)
# ax[0].set_title('Walk - X vs Time')
# ax[0].set_xlabel('Time (s)')
# ax[0].set_ylabel('X')

# ax[1].bar(walk['time'], walk['y'], width=0.5)
# ax[1].set_title('Walk - Y vs Time')
# ax[1].set_xlabel('Time (s)')
# ax[1].set_ylabel('Y')

# ax[2].bar(walk['time'], walk['z'], width=0.5)
# ax[2].set_title('Walk - Z vs Time')
# ax[2].set_xlabel('Time (s)')
# ax[2].set_ylabel('Z')

# # adjust layout and display the plots
# plt.tight_layout()
# plt.show()

```
